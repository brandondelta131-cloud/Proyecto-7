```python
st.title("Análisis de Datos de Vehículos Usados")
st.write("Este proyecto analiza un conjunto de datos de vehículos usados para identificar tendencias y patrones en precios y características")
```

```python
veiculos_df = pd.read_csv(r"C:\Users\chand\OneDrive\Documents\proyectos\sprint 7\vehicles_us.csv")
print(veiculos_df)
```

```python
#Visualización de los datos
print(veiculos_df.head(10))

#Análisis de dataset
print(veiculos_df.isnull().sum())

nan_rows = veiculos_df[veiculos_df.isnull().any(axis=1)]
print(nan_rows)
```

```python
#Limpieza de datos - Normalizar modelo
veiculos_df['model'] = veiculos_df['model'].str.lower()

#Rellenar valores ausentes
#Valores de columna is_4wd
veiculos_df['is_4wd'] = veiculos_df['is_4wd'].fillna(0)

#Valores de columna model_year
veiculos_df['model_year'] = veiculos_df.groupby('model')['model_year'].transform(lambda x: x.fillna(x.median()))

#Valores de columna cylinders
veiculos_df['cylinders'] = veiculos_df.groupby('model')['cylinders'].transform(lambda x: x.fillna(x.median()))

#Valores de la columna odometer
veiculos_df['odometer'] = veiculos_df.groupby('model_year')['odometer'].transform(lambda x: x.fillna(x.median()))

#Valores paint_color
veiculos_df['paint_color'] = veiculos_df['paint_color'].fillna('unknown')
```

```python
#Transformación de valores
veiculos_df['date_posted'] = pd.to_datetime(veiculos_df['date_posted'])
veiculos_df['month'] = veiculos_df['date_posted'].dt.month
veiculos_df['year_posted'] = veiculos_df['date_posted'].dt.year

veiculos_df['cylinders'] = veiculos_df['cylinders'].astype(int)
veiculos_df['model_year'] = veiculos_df['model_year'].astype(int)
veiculos_df['is_4wd'] = veiculos_df['is_4wd'].astype(int)

print(veiculos_df.info())
```

```python
#Análisis de outliers
plt.boxplot(veiculos_df['price'].dropna(), vert=False)
plt.title('Distribución de precios')
plt.ylabel('precio')
plt.show()

#Eliminación de outliers
q1 = veiculos_df['price'].quantile(0.25)
q3 = veiculos_df['price'].quantile(0.75)
IQR = q3 - q1
veiculos_df = veiculos_df[(veiculos_df['price'] >= q1 - 0.5*IQR) & (veiculos_df['price'] <= q3 + 0.5*IQR)]
```

```python
#Creación de nuevas variables
veiculos_df['age'] = 2024 - veiculos_df['model_year']
veiculos_df['price_per_mile'] = veiculos_df['price'] / veiculos_df['odometer']
veiculos_df['price_per_year'] = veiculos_df['price'] / veiculos_df['age']

print(veiculos_df)
```
